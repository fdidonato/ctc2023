```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('housesprices.csv',names=['serial','year','age','distance','stores','latitude','longitude','price'])
```

```python
dataset.info()
```

```python
dataset = dataset.drop(["serial"],axis=1);
dataset.head()
```

```python
X = dataset.drop('price',axis=1); # creiamo l'array numpy delle features droppando dal dataset l'ultima colonna che contiene le labels
Y = dataset['price'].values; # creiamo l'array numpy prendendo dal dataset solamente la colonna che contiene le labels
```

```python
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.1)
```

```python
model = LinearRegression()
history = model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
Y_pred
```
